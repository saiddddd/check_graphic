```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for plotting
np.random.seed(0)
thresholds = np.arange(1, 26)
or_de = np.cumsum(np.random.rand(25) * 10)
r_jade = np.cumsum(np.random.rand(25) * 9)
proposed_method = np.cumsum(np.random.rand(25) * 8.5)

generations = np.arange(1, 8001)
bcv_or_de = 3000 + np.cumsum(np.random.rand(8000) * 0.01)
bcv_r_jade = 3000 + np.cumsum(np.random.rand(8000) * 0.01)
bcv_proposed = 3000 + np.cumsum(np.random.rand(8000) * 0.01)

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(14, 6))

# Plot (a)
axs[0].plot(thresholds, or_de, label='OR-DE', color='blue')
axs[0].plot(thresholds, r_jade, label='R_j-JADE', color='black', linestyle='dotted')
axs[0].plot(thresholds, proposed_method, label='Proposed Method', color='red', linestyle='dashed')
axs[0].set_xlabel('number of thresholds')
axs[0].set_ylabel('The threshold values')
axs[0].legend()
axs[0].set_title('(a)')
axs[0].set_xlim(0, 25)
axs[0].set_ylim(0, max(or_de) + 10)

# Inset for plot (a)
inset_a = fig.add_axes([0.17, 0.5, 0.2, 0.3])
inset_a.plot(thresholds[3:9], or_de[3:9], color='blue')
inset_a.plot(thresholds[3:9], r_jade[3:9], color='black', linestyle='dotted')
inset_a.plot(thresholds[3:9], proposed_method[3:9], color='red', linestyle='dashed')
inset_a.set_xlim(4, 8)
inset_a.set_ylim(min(or_de[3:9]) - 2, max(or_de[3:9]) + 2)

# Plot (b)
axs[1].plot(generations, bcv_or_de, label='OR-DE', color='blue', linewidth=0.5)
axs[1].plot(generations, bcv_r_jade, label='R_j-JADE', color='black', linestyle='dotted', linewidth=0.5)
axs[1].plot(generations, bcv_proposed, label='Proposed Meth.', color='red', linestyle='dashed', linewidth=0.5)
axs[1].set_xlabel('Generations (t)')
axs[1].set_ylabel('Between-class variance (BCV)')
axs[1].legend()
axs[1].set_title('(b)')
axs[1].set_xlim(0, 8000)
axs[1].set_ylim(2996, max(bcv_or_de) + 2)

# Inset for plot (b)
inset_b = fig.add_axes([0.67, 0.5, 0.2, 0.3])
inset_b.plot(generations[7850:8000], bcv_or_de[7850:8000], color='blue', linewidth=0.5)
inset_b.plot(generations[7850:8000], bcv_r_jade[7850:8000], color='black', linestyle='dotted', linewidth=0.5)
inset_b.plot(generations[7850:8000], bcv_proposed[7850:8000], color='red', linestyle='dashed', linewidth=0.5)
inset_b.set_xlim(7850, 8000)
inset_b.set_ylim(min(bcv_or_de[7850:8000]) - 0.5, max(bcv_or_de[7850:8000]) + 0.5)

plt.tight_layout()
plt.show()
```
